```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import pdtypes

```

```python
df = pd.read_csv('data/Spotify_Song_Attributes.csv')
df.head()
```

```python
df.sort_values(by='msPlayed', ascending=False).head(20)
```

```python
tmp_df = (df
          .sort_values(by='msPlayed', ascending=False)
          .loc[:, ['trackName','artistName', 'msPlayed']]
          .assign(trackArtistName= lambda df: df['trackName'] + '(' + df['artistName'] + ')')
          .assign(sPlayed = lambda df: df['msPlayed']/1000/60/60)
          .head(20)
)
tmp_df
```

```python
sns.set_theme()
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(4,5))

g = sns.barplot(tmp_df, x='sPlayed', y='trackArtistName')
g.set_title('Sportipy 플레이 시간 높은 10개 곡')
g.set_xlabel("시간")
g.set_ylabel("곡명(가수명)")
g.bar_label(g.containers[0])

plt.show()
```

```python
tmp_df = (df
 .query(" genre == 'classical' ")
 .sort_values(by='msPlayed', ascending=False)
 .assign(sPlayed = lambda df: df['msPlayed']/1000/60/60)
 .assign(trackArtistName = lambda df: df['trackName'] + '(' + df['artistName'] + ')')
 .loc[:,['trackName','artistName','sPlayed','trackArtistName']]
 .head(20)
 )
```

```python
plt.pie(data=tmp_df, x='sPlayed', labels=tmp_df['trackArtistName'])
plt.show()
```
