```python
import matplotlib.pyplot as plt
import seaborn as sns
#시각화 한글 폰트 깨짐 한글 폰트 등록하기

plt.rc('font', family='D2Coding ligature') #예쁜 폰트 알아서 등록하면 됨.

```

```python
movie_df = movie_df.assign(release_time = lambda df:pd.to_datetime((df['release_time'])))
```

```python
movie_df.info()
```

```python
sns.set_theme()
plt.rc('font', family='Malgun Gothic') #set_theme() 이후 새로 바뀌므로 폰트를 다시 설정해주어야함.

plt.figure(figsize=(10,4))
g = sns.lineplot(data=movie_df, x='release_time', y='box_off_num')
g.axhline(10_000_000, color='red', linestyle='--')
g.set_xlabel('개봉일')
g.set_ylabel('관객수')
g.set_title('개봉일에 따른 관객수')
plt.show()
```

```python
#쿼리 질문 , 차트 보고 다시 질문(쿼리) EDM 통찰분석 : 1천만 관객 이상의 데이터를 새로 불러옴
movie_10k_df = movie_df.query("box_off_num > 10000000").reset_index(drop=True) # 과거의 인덱스를 새로 세팅하고 지우기도 함.
movie_10k_df
```

```python
#상영시간, 관람객 수의 관게
movie_df.loc[:,['time','box_off_num']]

plt.figure(figsize=(7,5))
g = sns.scatterplot(movie_df, x='time',y='box_off_num')
g.set_xlabel('상영시간(분)')
g.set_ylabel('관객수')
g.set_title('상영시간이 120~140분 사이에 관객수가 가장 많다.') # 타이틀에는 결과를 넣어주는게 좋은 시각화임. 시각화의 내용을 보여주는 결과값을 표현해주는게 좋다.
plt.show()
```

```python
movie_10k_df
```

```python
movie_10k_df = movie_10k_df.sort_values(by='box_off_num', ascending=False)

plt.figure(figsize=(4,3))
#sns.barplot(movie_10k_df, x='title', y='box_off_num') # x축 글자가 잘리면 x, y를 바꾸면됨.
g= sns.barplot(movie_10k_df, y='title', x='box_off_num')
g.bar_label(g.containers[0], labels=[f'{x:,.0f})' for x in g.containers[0].datavalues],
            label_type='edge', color='blue', fontsize=10)  #숫자 표시하는 것 붙여넣기 해서 씀.
g.set_title('천만관객 넘은 영화 순서')

plt.show()
```

```python
#py chart - matplotlib을 이용 : 비교가 어려움. 파이차트, 3D차트 (비추)
#장르별 빈도수 -> 그룹바이

tmp_df = movie_df.groupby('genre', as_index=False).size() # 빈도 agg아니고 size()

plt.figure(figsize=(4,4))
plt.pie(x=tmp_df['size'], labels=tmp_df['genre'], autopct='%.1f%%')
plt.show()
```
